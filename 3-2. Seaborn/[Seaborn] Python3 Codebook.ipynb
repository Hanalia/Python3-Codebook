## 🔎 Violin plot

- **sns .violinplot ( x = None , y = None , hue = None , data = None , order = None , hue_order = None , bw = ' scott ' , cut = 2 , scale = ' area ' ,  scale_hue = True , gridsize = 100 , width = 0.8 , inner = ' box ' , split = False , dodge = True , orient = None , linewidth = None , color = None , palette = None , saturation = 0.75 , ax = None , **kwargs)**: Violinplot은 Boxplot + Kernel Density Estimate 2개를 합친 plot이다. Boxplot과 달리 사각형이 아닌 각 data points를 나타낸다.

- Kernel Density Estimate : 관측 데이터 각각마다 중심으로 하는 커널 함수를 생성하고, 이렇게 만들어진 커널 함수들을 모두 더한 후 전체 데이터 개수로 나누어 하나의 밀도함수를 만드는 방법. 사용한 커널함수에 따라 모양이 달라진다. 히스토그램 확률밀도함수를 smoothing한 것으로 생각하면 이해하기 쉽다. 흰 색 점이 중위값이다. 실제 데이터의 전체 분포를 보여준다는 장점!!

### 📔 Examples

````
# sns.violinplot()
## Kernel Density function: 실제 데이터의 분포를 보여줌
sns.violinplot(data=titanic, x="age", y="sex", hue="survived")
````

## 🔎 Pair plot
